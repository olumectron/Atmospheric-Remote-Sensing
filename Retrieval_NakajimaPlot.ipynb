```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cloud1
headers = ["WLINF", "WSUP", "FFEW", "TOPDN", "TOPUP", "TOPDIR", "BOTDN", "BOTUP", "BOTDIR"]
df=pd.read_csv('alfa_out_new1',sep=" +",engine="python", names = headers)# Input file for cloud 1
reflect=df.values

szz=22.1 #for cloud 1
sza=szz*(np.pi/180)
solar_2p13=92.24782662
solar_0p65=1873.083526
pi=np.pi
ref0p86= (pi*reflect[4:630:10,0]/0.001)/(solar_0p65*np.cos(sza))
ref2p13= (pi*reflect[9:630:10,0]/0.001)/(solar_2p13*np.cos(sza))

def gtau(var,b):
    #Use this function to get the reflectances for the tau values which cannot be found through easy indexing
    tau = []
    for p in np.arange(0,7):
        tau.append(var[p::7][b])
    return np.array(tau)

#Plotting of diagram
fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)

reff = [5,6,16,20,25,32,64]
taus = [0,2,4,5,6,8,12,16,38]

for p in np.arange(0,7):
    ax1.plot((ref0p86)[p::7],(ref2p13)[p::7],color='r',ls='solid',lw=1)
    ax1.text((ref0p86)[p::7][-1]+.050, (ref2p13)[p::7][-1], r'r$_{eff}$=%s $\mu$m'%reff[p],\
             horizontalalignment='center',verticalalignment='center',fontsize=8,color='r')

for p in np.arange(0,9):
    ax1.plot(gtau((ref0p86),p),gtau((ref2p13),p),color='black',ls='dashed',lw=1)
    ax1.text(gtau((ref0p86),p)[-1], gtau((ref2p13),p)[-1]-.01, r'$\tau$=%s'%taus[p],\
             horizontalalignment='center',verticalalignment='center',fontsize=10,color='black')

ax1.scatter(0.4477,0.2613, label="Reflectance(Modis)", c="w",edgecolors='b', linewidths=5.0)
ax1.set_title(' Look up table (N-K diagram) For CLOUD 1  ',fontsize=14)#Plot title for cloud4
ax1.set_xlabel('VNIR Reflectance ($\lambda$=0.55 $\mu$m)',fontsize=14)
ax1.set_ylabel('SWNIR Reflection ($\lambda$=2.13 $\mu$m)',fontsize=14)

ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

ax1.tick_params(axis='y',labelsize=12)
ax1.tick_params(axis='x',labelsize=12)
ax1.minorticks_on()
ax1.legend(edgecolor='black')
ax1.grid(True, which='both')
plt.show()
```
