```python
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import pyplot
%matplotlib inline
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import math
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
```

```python
response = pd.read_csv('tau.csv')
pd.set_option('display.max_columns', None)
response.head(3)
```

```python
residual = response['Retrieval_tau']-response['Modis_tau']
residual_mean = response['Retrieval_tau'] - response['Modis_tau'].mean()
mse = np.sum(np.square(residual))/len(residual)
rmse = np.sqrt(np.sum(np.square(residual))/len(residual))
mae = np.sum(np.abs(residual))/len(residual)
rsquare = 1 - (np.sum(np.square(residual))/np.sum(np.square(residual_mean)))
print(f'MSE = {mse}')
print(f'RMSE = {rmse}')
print(f'MAE = {mae}')
print(f'R2 = {rsquare}')
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style('whitegrid')
ax = sns.regplot(x="Modis_tau", y="Retrieval_tau", data=response, scatter_kws={"color": "blue", "alpha": 0.7}, line_kws={"color": "black"})

# Set the graph's title
ax.set_title("Scatter plot of retrieved Cloud optical depth" ,fontsize= 16, weight='semibold')
ax.set_xlabel("Modis(Cloud optical depth)" ,fontsize= 14)
ax.legend()
ax.set_ylabel("Retrieved(Cloud optical depth)" ,fontsize= 14)

plt.show()
```
